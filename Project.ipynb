```python
# Write a program to count Digits, Letters and Spaces in a String.
def count_chars(string):
    digit_count = 0
    letter_count = 0
    space_count = 0

    for char in string:
        if char.isdigit():
            digit_count += 1
        elif char.isalpha():
            letter_count += 1
        elif char.isspace():
            space_count += 1

    return digit_count, letter_count, space_count

# Test the function
input_string = input("Enter a string: ")
digit_count, letter_count, space_count = count_chars(input_string)
print("Digits:", digit_count)
print("Letters:", letter_count)
print("Spaces:",space_count)
```

```python
# Write a Python program to get the sum of a non-negative integer using recursion.
# Test Data:
#                   	sumDigits(345) -> 12
#                   	sumDigits(45) -> 9
def sumDigits(n):
    if n == 0:
        return 0
    else:
        return n % 10 + sumDigits(n // 10)

# Test the function
print("sumDigits(345) ->", sumDigits(345))
print("sumDigits(45) ->",sumDigits(45))
```

    What is the difference between Python Arrays and lists?

Definition and Purpose:

Lists: Lists are a built-in data structure in Python. They can hold elements of different data types. Lists are dynamic (size can change during runtime). Ideal for shorter sequences of data items.

Arrays: Arrays must be imported from the array or numpy package. They hold elements of the same data type (homogeneous). Arrays have a fixed size (typically). Preferred for longer sequences of data items.

Memory Allocation:

Lists: Lists are implemented as dynamic arrays. Memory allocation is flexible.

Arrays: Arrays allocate contiguous memory locations. Faster indexing due to contiguous memory.

Insertion and Deletion:

Lists: Insertion and deletion can be less efficient (especially at the beginning or middle). Requires shifting subsequent elements.

Arrays: Insertion and deletion are expensive. All elements after the insertion/deletion point must be shifted.

Data Types:

Lists: Can store elements of different data types.

Arrays: Store elements of the same data type.

Import Requirement:

Lists: Built-in; no need for explicit import.

Arrays: Must explicitly import the array module.

Arithmetic Operations:

Lists: Cannot directly handle arithmetic operations.

Arrays: Can directly handle arithmetic operations.

    What is a dictionary in Python?

In Python, a dictionary is an unordered collection of data in a key-value pair format. It is also known as a hash table or associative array in other programming languages. Dictionaries are mutable, which means they can be modified after creation. Each key in a dictionary must be unique and immutable (such as strings, numbers, or tuples), while the corresponding value can be of any data type and mutable.Dictionaries are denoted by curly braces {}, and the key-value pairs are separated by a colon :. Here's a basic example of a dictionary in Python:

     Explain split(), sub(), subn() methods of “re” module in Python .

The re module in Python provides several methods for working with regular expressions. Among these methods are split(), sub(), and subn(), which are used for splitting strings and replacing substrings based on patterns. Here's an explanation of each:

split(pattern, string, maxsplit=0):

The split() method splits a string into a list of substrings based on a specified pattern (regular expression).
It takes three parameters:
pattern: The regular expression pattern to search for.
string: The string to be split.
maxsplit (optional): Specifies the maximum number of splits. If omitted or zero, there is no limit on the number of splits.
The method returns a list of substrings resulting from the split operation.
Example:

```python
import re
text = "apple, banana, cherry"
result = re.split(",\s*", text)
print(result)
```



sub(pattern, repl, string, count=0):

The sub() method replaces occurrences of a pattern in a string with a specified replacement string.
It takes four parameters:
pattern: The regular expression pattern to search for.
repl: The replacement string to substitute for matches.
string: The input string to search and perform replacements on.
count (optional): Specifies the maximum number of replacements to make. If omitted or zero, all occurrences will be replaced.
The method returns a new string with replacements applied.
Example:

```python
import re
text = "apple, banana, cherry"
result = re.sub(r'\b\w+\b', 'fruit', text)
print(result)
```

subn(pattern, repl, string, count=0):

The subn() method is similar to sub(), but it also returns the number of replacements made along with the modified string.
It takes the same parameters as sub().
The method returns a tuple containing the modified string and the count of replacements made.
Example:

```python
import re
text = "apple, banana, cherry"
result, count = re.subn(r'\b\w+\b', 'fruit', text)
print(result)
print(count)
```

    Explain Inheritance in Python with an example.

Inheritance is a fundamental concept in object-oriented programming (OOP) where a new class (subclass) can inherit attributes and methods from an existing class (superclass). This allows for code reuse and promotes the creation of hierarchical relationships between classes. In Python, inheritance is implemented by specifying the superclass(es) in the definition of a subclass.

Here's an example of inheritance in Python:


```python
# Define the superclass (parent class)
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")

# Define a subclass (child class) inheriting from Animal
class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

# Define another subclass (child class) inheriting from Animal
class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

# Create instances of subclasses
dog = Dog("Buddy")
cat = Cat("Whiskers")

# Call methods inherited from the superclass
print(dog.speak())
print(cat.speak())
```

    Difference between Supervised and Unsupervised learning.

The key differences between Supervised and Unsupervised learning:

Supervised Learning:
Definition:
In supervised learning, models are trained using labeled data.
Labeled data means each input example is paired with its corresponding output (target).
Objective:
The goal is to find a mapping function that can predict the output (target) for new, unseen data.
Examples:
Classification: Assigning input data to predefined categories (e.g., spam or not spam).
Regression: Predicting a continuous output (e.g., house prices based on features).
Feedback:
Supervised learning models receive direct feedback during training.
Similar to how a student learns with a teacher’s guidance.
Unsupervised Learning:
Definition:
In unsupervised learning, models work with unlabeled data.
There are no predefined output labels.
Objective:
The goal is to find patterns, structures, or relationships within the data.
Examples:
Clustering: Grouping similar data points together (e.g., customer segmentation).
Dimensionality Reduction: Reducing the number of features while preserving information.
Association: Discovering relationships between items (e.g., market basket analysis).
Feedback:
Unsupervised learning models do not receive direct feedback.
They explore data independently, without predefined answers.
Comparison:
Supervised Learning:
Requires labeled data.
Predicts output.
Needs direct supervision.
Unsupervised Learning:
Uses unlabeled data.
Finds hidden patterns.
Works without supervision.
